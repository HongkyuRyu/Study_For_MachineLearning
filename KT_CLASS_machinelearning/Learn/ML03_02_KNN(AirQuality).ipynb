# 1.환경 준비

- 기본 라이브러리와 대상 데이터를 가져와 이후 과정을 준비합니다.

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/airquality_simple.csv'
data = pd.read_csv(path)
```

# 2.데이터 이해

- 분석할 데이터를 충분히 이해할 수 있도록 다양한 탐색 과정을 수행합니다.

# 3.데이터 준비

- 전처리 과정을 통해 머신러닝 알고리즘에 사용할 수 있는 형태의 데이터를 준비합니다.

**1) 결측치 처리**

- 결측치가 있으면 제거하거나 적절한 값으로 채웁니다.
- 시계열 데이터이므로 선형보간법으로 채웁니다.

```python
# 결측치 확인
data.isnull().sum()
```

```python
# 결측치 채우기
data.interpolate(method='linear', inplace=True)

# 확인
data.isnull().sum()
```

**2) 변수 제거**

- 분석에 의미가 없다고 판단되는 변수는 제거합니다.
- Month, Day 열을 제거합니다.

```python
#  변수 제거
drop_cols = ['Month', 'Day']
data.drop(drop_cols, axis=1, inplace=True)

# 확인
data.head()
```

**3) x, y 분리**

- target 변수를 명확히 지정합니다.
- target을 제외한 변수를 x 데이터프레임으로 선언합니다.
- y 데이터프레임은 target 변수만을 가집니다.

```python
# target 확인
target = 'Ozone'

# 데이터 분리
x = data.drop(target, axis=1)
y = data.loc[:, target]
```

**4) 학습용, 평가용 데이터 분리**

- 학습용, 평가용 데이터를 적절한 비율로 분리합니다.
- 반복 실행 시 동일한 결과를 얻기 위해 random_state 옵션을 지정합니다.

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 데이터 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)  
```

**4) 정규화**

- KNN 알고리즘을 사용하기 위해 정규화를 진행합니다.
- 다음 중 한 가지 방법을 사용해 진행합니다.

- 방법1: 함수 사용 -> 배열로 바뀜. (column들이 모두 날라감)

```python
print(x_train.max(), x_test.max())
# train데이터의 최대값, 최소값 으로 test데이터를 정규화.
# 미래의 데이터가 현재 데이터에 영향을 주면 안됨.
# 그래서, 모든 것은 학습데이터를 가지고 해야한다. (학습 데이터를 가지고 평가데이터 정규화를 실행한다,)
# 나중에 test데이터셋이 주어질때 , validation set도 마찬가지. (학습 데이터를 가지고 검증데이터에 대해 정규화를 실행한다.)
# 모듈 불러오기
from sklearn.preprocessing import MinMaxScaler

# 정규화
x_col = list(x)
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

x_train = pd.DataFrame(x_train, columns=x_col)
x_test = pd.DataFrame(x_test, columns=x_col)
```

```python
x_train.head()
```

- 방법2: 공식 사용 -> 여전히 dataFrame으로 남음.

$$ \huge x_{norm}=\frac{x-x_{min}}{x_{max}-x_{min}} $$

```python
# 최댓값, 최솟값 구하기
x_max = x_train.max()
x_min = x_train.min()

# 정규화
x_train = (x_train - x_min)/(x_max - x_min)
x_test = (x_test - x_min) / (x_max - x_min)
```

```python
x_train.head()
```

# 4.모델링

- 본격적으로 모델을 선언하고 학습하고 평가하는 과정을 진행합니다.

```python
# 1단계: 불러오기
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# 2단계: 선언하기
model = KNeighborsRegressor(n_neighbors=5)
```

```python
# 3단계: 학습하기
model.fit(x_train, y_train)
```

```python
# 4단계 예측하기
y_pred = model.predict(x_test)
```

```python
# 5단계: 평가하기

print('MAE: ', mean_absolute_error(y_test, y_pred))
print('훈련데이터:', model.score(x_train, y_train))
print('테스트데이터: ' , r2_score(y_test, y_pred))
```

# 5.기타

- 기타 필요한 내용이 있으면 진행합니다.

```python
# 예측값, 실젯값 시각화
plt.plot(y_test.values, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.ylabel('Ozone')
plt.show()
```
