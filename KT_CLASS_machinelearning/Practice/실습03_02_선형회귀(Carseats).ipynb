# 1.환경 준비

- 기본 라이브러리와 대상 데이터를 가져와 이후 과정을 준비합니다.

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/Carseats.csv'
data = pd.read_csv(path)
```

# 2.데이터 이해

- 분석할 데이터를 충분히 이해할 수 있도록 다양한 탐색 과정을 수행합니다.

```python
# 상위 몇 개 행 확인
data.head()
```

**데이터 설명**

- Sales: 각 지역 판매량(단위: 1,000개)
- CompPrice: 경쟁사 가격 (단위: 달러)
- Income: 지역 평균 소득 (단위: 1,000달러)
- Advertising: 각 지역, 회사의 광고 예산 (단위: 1,000달러)
- Population: 지역 인구 수 (단위: 1,000명)
- Price: 자사 지역별 판매 가격 (단위: 달러)
- ShelveLoc: 진열 상태
- Age: 지역 인구의 평균 연령
- Education: 각 지역 교육 수준
- Urban: 도심 지역 여부 (Yes,No)
- US: 매장이 미국에 있는지 여부 (Yes, No)

# 3.데이터 준비

- 전처리 과정을 통해 머신러닝 알고리즘에 사용할 수 있는 형태의 데이터를 준비합니다.

**1) x, y 분리**

```python
# target 확인
target = 'Sales'

# 데이터 분리
x = data.drop(target, axis=1)
y = data.loc[:, target]
```

**2) 가변수화**

```python
# 가변수화 대상: ShelveLoc, Education, Urban, US
dumm_cols = ['ShelveLoc', 'Education', 'Urban', 'US']

# x 데이터프레임의 대한 가변수화
x = pd.get_dummies(x, columns=dumm_cols, drop_first=True)

# 확인
x.head()
```

**3) 학습용, 평가용 데이터 분리**

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 데이터 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

# 4.모델링

- 본격적으로 모델을 선언하고 학습하고 평가하는 과정을 진행합니다.

```python
# 1단계: 불러오기
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, ElasticNetCV
```

```python
# 2단계: 선언하기
model = LinearRegression()
RidgeModel = RidgeCV(alphas=np.linspace(0.1, 30, 20), cv=5)
LassoModel = LassoCV(alphas=np.linspace(0.1, 10, 50), cv=5, random_state=1)
ElasticModel = ElasticNetCV(l1_ratio=np.linspace(0.1, 1, 10), alphas=np.linspace(0.1, 10, 50), cv=5, random_state=1)
```

```python
# 3단계: 학습하기
m1 = model.fit(x_train, y_train)
m2 = RidgeModel.fit(x_train, y_train)
m3 = LassoModel.fit(x_train, y_train)
m4 = ElasticModel.fit(x_train, y_train)
```

```python
# 회귀계수 확인
print(list(x))

print(m1.coef_)
print('--'*10)
print(m2.coef_)
print('--'*10)
print(m3.coef_)
print('--'*10)
print(m4.coef_)
```

```python
# 회귀계수 시각화
temp = pd.DataFrame({
    'feature': list(x),
    'weight': m1.coef_
})
temp.head()
```

```python
plt.barh(temp['feature'], temp['weight'])
plt.show()
```

```python
temp.sort_values(by='weight', ascending=True, inplace=True)

plt.barh(temp['feature'], temp['weight'])
plt.show()
```

```python
# 나머지도 모두 식닥화
temp2 = pd.DataFrame({
    'feature': list(x),
    'weight': m2.coef_
})
temp3 = pd.DataFrame({
    'feature': list(x),
    'weight': m3.coef_
})
temp4 = pd.DataFrame({
    'feature': list(x),
    'weight': m4.coef_
})
temp2.sort_values(by='weight', ascending=True, inplace=True)
temp3.sort_values(by='weight', ascending=True, inplace=True)
temp4.sort_values(by='weight', ascending=True, inplace=True)

plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.barh(temp['feature'], temp['weight'], label='Linear')
plt.legend()
plt.subplot(2, 2, 2)
plt.barh(temp2['feature'], temp2['weight'], label='Ridge')
plt.legend()
plt.subplot(2, 2, 3)
plt.barh(temp3['feature'], temp3['weight'], label='Lasso')
plt.legend()
plt.subplot(2, 2, 4)
plt.barh(temp4['feature'], temp4['weight'], label='Elastic')
plt.legend()
plt.show()
```

```python
# 4단계: 예측하기
y_pred1 = m1.predict(x_test)
y_pred2 = m2.predict(x_test)
y_pred3 = m3.predict(x_test)
y_pred4 = m4.predict(x_test)
```

```python
# 5단계: 평가하기
from sklearn.metrics import r2_score

print('Linear 학습 성능: ', m1.score(x_train, y_train))
print('Linear 평가 성능: ', r2_score(y_test, y_pred1))
print('--'*10)
print('Ridge 학습 성능: ', m2.score(x_train, y_train))
print('Ridge 평가 성능: ', r2_score(y_test, y_pred2))
print('--'*10)
print('Lasso 학습 성능: ', m3.score(x_train, y_train))
print('Lasso 평가 성능: ', r2_score(y_test, y_pred3))
print('--'*10)
print('Elastic 학습 성능: ', m4.score(x_train, y_train))
print('Elastic 평가 성능: ', r2_score(y_test, y_pred4))
```
